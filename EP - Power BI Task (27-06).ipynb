## Data Cleaning

```python
# Percentage of missing values
(df.isnull().sum() / len(df)) * 100
```

#### Cleaning cargo_volume_l Column

```python
banana_enteries = df['cargo_volume_l'].str.contains('Banana Boxes', na=False)
```

```python
df.loc[banana_enteries, 'cargo_volume_l'] = np.nan
```

```python
df['cargo_volume_l'] = pd.to_numeric(df['cargo_volume_l'], errors='coerce')
```

#### Imputing Missing Values

```python
num_cols = df.select_dtypes(include='number')
cat_cols = df.select_dtypes(include='object')

for col in num_cols:
    df[col] = df[col].fillna(df[col].median())
for col in cat_cols:
    df[col] = df[col].fillna(df[col].mode()[0])
```

```python
# Checking data types and nulls again
print(df.dtypes, '\n')
print(df.isnull().sum())
```

## Exploratory Data Analysis (EDA)

#### Correlation Analysis (Range vs Features)

```python
# Selecting numeric columns to compare with range
num_cols = ["range_km", "battery_capacity_kWh", "efficiency_wh_per_km", 
            "fast_charging_power_kw_dc", "acceleration_0_100_s", "top_speed_kmh"]

# Ploting correlation heatmap
plt.figure(figsize=(10,6))
sns.heatmap(df[num_cols].corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Matrix: Range vs Key Features")
plt.show()
```

#### Scatter Plots â€“ Relationship with Range

```python
# Range vs Battery Capacity
sns.scatterplot(data=df, x="battery_capacity_kWh", y="range_km", hue="brand")
plt.title("Range vs Battery Capacity")
plt.show()

# Range vs Efficiency
sns.scatterplot(data=df, x="efficiency_wh_per_km", y="range_km")
plt.title("Range vs Efficiency (Lower = Better)")
plt.show()
```

## Average Acceleration & Top Speed by Body Type

```python
# Group by body type
grouped = df.groupby("car_body_type")[["acceleration_0_100_s", "top_speed_kmh"]].mean().sort_values(by="top_speed_kmh", ascending=False)

# Barplot
grouped.plot(kind="bar", figsize=(10,5))
plt.title("Avg Acceleration and Top Speed by Body Type")
plt.ylabel("Value")
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

#### Charging Speed vs Battery Size / Range

```python
# Fast charging vs battery
sns.scatterplot(data=df, x="fast_charging_power_kw_dc", y="battery_capacity_kWh")
plt.title("Fast Charging Power vs Battery Size")
plt.show()

# Fast charging vs range
sns.scatterplot(data=df, x="fast_charging_power_kw_dc", y="range_km")
plt.title("Fast Charging Power vs Range")
plt.show()
```

#### Battery Size by Brand (Boxplot)

```python
top_brands = df["brand"].value_counts().head(10).index
filtered = df[df["brand"].isin(top_brands)]

plt.figure(figsize=(12,6))
sns.boxplot(data=filtered, x="brand", y="battery_capacity_kWh")
plt.title("Battery Capacity Distribution by Top 10 Brands")
plt.xticks(rotation=45)
plt.show()
```
